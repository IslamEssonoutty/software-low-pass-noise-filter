```python
import numpy as np
import matplotlib.pyplot as plt



```

```python
#low-pass windowed-sinc function
def low_pass_windowed_Generator(h):
    for i in range(98):
        h[i] = 0.31752 * np.sin(0.314159 * (i-49.00001)) / (i-49.00001)
        h[i] = h[i] * (0.54 - 0.46 * np.cos(0.0641114 * i))
 
    return h
```

```python
h=np.ones(98).tolist()
h=low_pass_windowed_Generator(h)
# Plot the filter kernel h[n]
plt.plot(h)
plt.title('Filter Kernel h[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()
```

```python
signal=np.ones(500).tolist()
for i in range(500):
    if i==0:
        signal[i] = 1
    else:
        signal[i] = 0
        
Y1=np.convolve(signal, h, mode='full')
#print('Output of convolution: y[n] is a filtered version of x[n] that passes sinusoids with fewer than 25 cycles in 500 samples, and blocks higher frequency sinusoids.')
#print('Explanation: Since the input signal x[n] is an impulse, the output y[n] will be equal to the filter kernel h[n]. Therefore, y[n] will contain the low-pass characteristics of h[n], passing only those frequencies that have fewer than 25 cycles in 500 samples.')

plt.plot(Y1)
```

```python
n = np.arange(500)
y = 1*np.sin(2*np.pi*6*n/500) + 0.5*np.sin(2*np.pi*44*n/500)




plt.plot(y)
plt.title('Test Signal y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()
```

```python
y_after_filter=np.convolve(y, h, mode='full')
plt.plot(y_after_filter)
plt.title('Result Signal y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()
```
